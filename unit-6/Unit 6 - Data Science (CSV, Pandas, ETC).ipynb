## - The Index
 -- The index represents the column.  For example:
     -- row[0], row[1] represent the ProductID and Name columns.  The for loop will then iterate through the entire collection, printing the value for every row's 0 and 1 index

```python
with open('products-2.csv') as csvfile:
    readCSV = csv.reader(csvfile, delimiter='\t')
    # print(readCSV)
    product_ids = []
    names = []
    
    for row in readCSV:
        product_id = row[0]
        name = row[1]
        
        product_ids.append(product_id)
        names.append(name)
    
print(product_ids)
print(names)
```

```python
with open('products-2.csv') as csvfile:
    readCSV = csv.reader(csvfile, delimiter='\t')
    # print(readCSV)
    product_ids = []
    names = []
    
    for row in readCSV:
        product_id = row[0]
        name = row[1]
        
        product_ids.append(product_id)
        names.append(name)
    
# print(product_ids)
# print(names)

where_is_product_id = input('Where is this product ID?')
prod_index = product_ids.index(where_is_product_id)
print(prod_index)
```
